SIMPLE LINEAR REGRESSION

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
# load dataset
path = "../dataset/hour.csv"
df = pd.read_csv(path)

print("Dataset shape:", df.shape)
```

```python
# using temperature as predictor
X = df[['temp']]
y = df['cnt']
```

```python
# split data (different from your friend)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=7
)
```

```python
# scatter plot
plt.scatter(X_train, y_train, alpha=0.4)
plt.xlabel("Temperature")
plt.ylabel("Bike Count")
plt.title("Temperature vs Bike Rentals")
plt.show()

```

```python
model = LinearRegression()
model.fit(X_train, y_train)

print("Intercept:", model.intercept_)
print("Coefficient:", model.coef_[0])

```

```python
# prediction on test data
y_pred = model.predict(X_test)
```

```python
# evaluation metrics
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print("R2 Score:", round(r2, 4))
print("Mean Squared Error:", round(mse, 2))
```

```python
# regression line plot
sorted_idx = np.argsort(X_test.values.flatten())

plt.scatter(X_test, y_test, alpha=0.3, label="Actual")
plt.plot(
    X_test.values.flatten()[sorted_idx],
    y_pred[sorted_idx],
    label="Regression Line"
)

plt.xlabel("Temperature")
plt.ylabel("Bike Rentals")
plt.title("Simple Linear Regression")
plt.legend()
plt.show()
```
